```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as  sns
```

```python
df.columns
```

```python
sns.relplot(x="year", y="totalprod", kind="line", data= df)
```

```python
sns.relplot(x="priceperlb", y="state", size="year", data= df, height=15, aspect=.5)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=df['totalprod'], y=df['numcol'], data= df, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('Honey Production Year over Year');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=df['totalprod'], y=df['yieldpercol'], ax=ax2, color="r", 
                         label ='Yield per colony', legend=False) 
sns.despine(right=False)
plt.ylabel('colony yield')
ax.figure.legend()
```
