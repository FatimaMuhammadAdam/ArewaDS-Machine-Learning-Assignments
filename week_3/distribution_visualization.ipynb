```python
#showing the smoker count
sns.countplot( x = 'smoker', data = df)
plt.title("The number of smoker ")
```

```python
#displaying the gender count
sns.countplot(x = 'sex', data = df)
plt.title("The gender distribution")
```

```python
#ploting the histogram chart for the body mass index(bmi)
sns.histplot ( x = df [ "bmi"], bins = 30)
plt.title("The distribution of Bmi (Body Mass Index)")
plt.show()
```

```python
#relationship between smoker and charges 
sns.boxplot(x = 'smoker', y = 'charges', data = df)
plt.title("The relationship between “smoker” and “charges")
```

```python
sns.kdeplot(df['bmi'])
plt.show()
```

```python


# Select the columns for the density plot
columns = ['children', 'charges']

# Create subplots for each column
fig, axes = plt.subplots(nrows=1, ncols=len(columns), figsize=(12, 4))

# Iterate over the columns and plot the density
for i, col in enumerate(columns):
    sns.kdeplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(col)

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
