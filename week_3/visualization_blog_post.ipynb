**A lollipop chart** is a type of data visualization that combines elements of a scatter plot and a bar chart. It is also known as a dot plot or a barbell chart. It is used to show and compare values for various categories or groups. In a lollipop chart, data points are represented by individual dots or circles, similar to a scatter plot.Some important parts of the lollipop chart are:
##
Lollipop charts are useful for comparing values across different categories or groups and highlighting differences or patterns.
##
They are particularly effective when you have numerical or quantitative data and want to display the magnitude or value associated with each category.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
ADS_ladies = ['Farida', 'Fatima', 'Maryam', 'Deborah', 'Rahma', 'zarau']
puntuality = [10, 20, 15.7, 9, 33.3, 44.9]

# Plotting
plt.stem(ADS_ladies, puntuality, basefmt=' ')
plt.xticks(ADS_ladies)
plt.ylabel('Values')
plt.title('Puntuality level of ArewaDS ladies')

# Display
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the data
ADS_ladies = ['Farida', 'Fatima', 'Maryam', 'Deborah', 'Rahma', 'Zarau']
Participating_in_office_hours = ['Saturday', 'Sunday', 'Tuesday', 'Wednesday', 'Friday', 'Monday']

# Create a list to represent the heatmap data that 6x6 matrics of random numbers 
data = [[1, 0, 1, 0, 4, 1],
        [1, 0, 3, 0, 1, 0],
        [0, 1, 0, 1, 0, 0],
        [0, 1, 0, 2, 0, 1],
        [1, 0, 1, 0, 0, 3],
        [0, 2, 1, 1, 0, 0]]

# Create a heatmap using seaborn
sns.heatmap(data, cmap='YlGnBu', annot=True, fmt='d', xticklabels=Participating_in_office_hours, yticklabels=ADS_ladies)

# Add labels and title
plt.xlabel('Participating in Office Hours')
plt.ylabel('ADS Ladies')
plt.title('ADS Ladies Participation in Office Hours')

# Display the heatmap
plt.show()
```

The sns.heatmap() function is used to create the heatmap. The data parameter is passed to specify the data for the heatmap. The cmap parameter sets the color map for the heatmap (in this case, 'YlGnBu' is used). The annot parameter is set to True to display the numeric values in each cell of the heatmap. The fmt parameter specifies the format of the numbers displayed in the cells (in this case, 'd' represents integers). Want to learn more about heatmaps? No problem, just check the heatmap.

NOTE: When you're dealing with a large dataset in which you have read its file, whether in csv or any other format, all you need to do is pass or call the column names you want to show the relationship, so the matrix set of data is no longer needed, as shown in the example above.

**Line Chart:** A line chart displays data points connected by lines. It is useful for showing trends and changes in data over time.

```python
ADS_ladies = ['Farida', 'Fatima', 'Maryam', 'Deborah', 'Rahma', 'Zarau']
punctuality = [57, 60, 59, 45, 33.3, 44.9]

plt.plot(ADS_ladies, puntuality, linestyle = '-', marker = '*', color ='r')
plt.xlabel('ADS_ladies')
plt.ylabel('punctuality')
plt.title('puntuality')

plt.show()
```

**Histogram:** A histogram displays the distribution of a continuous variable by dividing it into bins and showing the frequency or count of values within each bin.

```python
import matplotlib.pyplot as plt

# Define the data
ADS_ladies = ['Farida', 'Fatima', 'Maryam', 'Deborah', 'Rahma', 'Zarau']
punctuality = [55.7, 60.9, 59, 45, 33.3, 44.9]

# Create a histogram using matplotlib
plt.hist(punctuality, bins=10, edgecolor='red')

# Add labels and title
plt.xlabel('Punctuality')
plt.ylabel('Frequency')
plt.title('Punctuality Distribution of ADS Ladies')

# Add x-axis labels
plt.xticks(range(30, 70, 5))

# Display the histogram
plt.show()
```

The plt.hist() function from the matplotlib library is used to create the histogram. The punctuality list is passed as the data for the histogram, and the bins parameter is set to 10 to specify the number of bins or intervals for the histogram. The edgecolor parameter sets the color of the edges of the bars.

**Area chart:** An area chart is similar to a line chart, but the area between the line and the x-axis is colored. It can be used to display cumulative data or to compare multiple data series.

```python
import matplotlib.pyplot as plt

# Define the data
ADS_ladies = ['Farida', 'Fatima', 'Maryam', 'Deborah', 'Rahma', 'Zarau']
punctuality = [45, 33.3, 57, 60, 59,  44.9]

# Create the area chart
plt.fill_between(range(len(ADS_ladies)), punctuality, color='yellow', alpha=0.4)

# Add labels and title
plt.xlabel('ADS Ladies')
plt.ylabel('Punctuality (%)')
plt.title('Punctuality of ADS Ladies')

# Customize the x-axis labels
plt.xticks(range(len(ADS_ladies)), ADS_ladies)

# Display the area chart
plt.show()
```

The fill_between() function from the plt module is used to create the area chart by filling the area between the data points and the x-axis.

The color parameter sets the color of the area, and the alpha parameter controls the transparency of the area. In this example, the area is filled with a yellow color and has an alpha value of 0.4. Click this link for more information on the area chart.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
ADS_ladies = ['Farida', 'Fatima', 'Maryam', 'Deborah', 'Rahma', 'Zarau']
punctuality = [57, 60, 59, 45, 33.3, 44.9]

# Create a figure and axes
fig, ax = plt.subplots()

# Set the y-axis range
ax.set_ylim([0, 100])

# Create horizontal bars for punctuality
y_pos = np.arange(len(ADS_ladies))
ax.barh(y_pos, punctuality, align='center', color='blue')

# Set y-axis ticks and labels
ax.set_yticks(y_pos)
ax.set_yticklabels(ADS_ladies)

# Set x-axis label
ax.set_xlabel('Punctuality')

# Add data labels to the bars
for i, v in enumerate(punctuality):
    ax.text(v + 1, i, str(v), color='black')

# Set title
ax.set_title('ADS Ladies Punctuality')

# Display the chart
plt.show()
```
