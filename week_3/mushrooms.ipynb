```python
import pandas as pd
import matplotlib.pyplot as plt
mushrooms = pd.read_csv('mushrooms.csv')
mushrooms.head()
```

```python
print(mushrooms.select_dtypes(["object"]).columns)
```

```python
cols = mushrooms.select_dtypes(["object"]).columns
mushrooms[cols] = mushrooms[cols].astype('category')
```

```python
edibleclass=mushrooms.groupby(['class']).count()
edibleclass
```

```python
mushrooms.head()
```

```python
labels=['Edible','Poisonous']
plt.pie(edibleclass['population'],labels=labels,autopct='%.1f %%')
plt.title('Edible?')
plt.show()
```

**Donuts**

Donut charts, also known as doughnut charts, are a type of circular visualization that displays data in the form of a ring. They resemble pie charts but have a central hole that gives them their distinct appearance.

```python
habitat=mushrooms.groupby(['habitat']).count()
habitat
```

```python
#displying the pie chart
labels=['Paths','Urban','Waste','Wood','Grasses','Leaves','Meadows']

plt.pie(habitat['class'], labels=labels,
        autopct='%1.1f%%', pctdistance=0.85) 
center_circle = plt.Circle((0, 0), 0.50, fc='white')
fig = plt.gcf()

fig.gca().add_artist(center_circle)
  
plt.title('Mushroom Habitats')
  
plt.show()
```

**Waffles!**

A 'waffle' chart, also known as a square pie chart, is a visualization technique used to represent parts of a whole. It is similar to a traditional pie chart but uses squares or rectangles instead of pie slices. Each square or rectangle in the chart represents a portion or category of the total data.

**OR**

A 'waffle' chart is a unique way to represent quantities as a 2D array of squares. Consider visualizing the various quantities of mushroom cap colors in this dataset.

```python
from pywaffle import Waffle
```

```python
capcolor=mushrooms.groupby(['cap-color']).count()
capcolor
```

```python
  
data ={'color': ['brown', 'buff', 'cinnamon', 'red', 'white', 'yellow', 'green', 'pink', 'purple'],
    'amount': capcolor['class']
     }
  
df = pd.DataFrame(data)
  
fig = plt.figure(
    FigureClass = Waffle,
    rows = 100,
    values = df.amount,
    labels = list(df.color),
    figsize = (30,30),
    colors=["brown", "tan", "maroon", "green", "pink", "purple", "blue", "whitesmoke", "yellow"],)
```
