**Preprocessing**

Let get info for the whole country in both the three different dataset using groupby :

```python
infected = infected.groupby('Country/Region').sum()
recovered = recovered.groupby('Country/Region').sum()
deaths = deaths.groupby('Country/Region').sum()

#Let display
infected.head()
```

```python
recovered.head()

```

```python
deaths.head()
```

Because of the use of groupby, all DataFrames are now indexed by Country/Region. We can thus use to access data for a specific country.loc:

```python
infected.loc['US'][2:].plot()
recovered.loc['US'][2:].plot()
plt.show()
```

```python
#LET CHECK for another country 
infected.loc['Angola'][2:].plot()
recovered.loc['Angola'][2:].plot()
plt.show()
```

Take **note** of how we use [2:] to remove the first two elements of a sequence that contain a country's geolocation. We can also remove those two columns entirely.

```python
infected.drop(columns=['Lat','Long'],inplace=True)
recovered.drop(columns=['Lat','Long'],inplace=True)
deaths.drop(columns=['Lat','Long'],inplace=True)
```

**Investigating the Data**

Let's move on to looking into a specific country. Let's make a frame that contains infection data indexed by date:

```python
def mkframe(country):
    df = pd.DataFrame({ 'infected' : infected.loc[country] ,
                        'recovered' : recovered.loc[country],
                        'deaths' : deaths.loc[country]})
    df.index = pd.to_datetime(df.index)
    return df

df = mkframe('US')
df
```

```python
df.plot()
plt.show()
```

Let us now estimate the number of new infected people each day. This will allow us to see how quickly the pandemic spreads. The simplest method is to use diff:

```python
df['ninfected'] = df['infected'].diff()
df['ninfected'].plot()
plt.show()
```

We can see high fluctuations in data. Let's look closer at one of the months:

```python
df[(df.index.year == 2020)& (df.index.month == 7)]['ninfected'].plot()
plt.show()

```

Because we want to see the trends, it makes sense to smooth out the curve by computing running average (i.e. for each day, we will compute the mean value of the previous a few days):

```python
df['ninfav'] = df['ninfected'].rolling(window=7).mean()
df['ninfav'].plot()
plt.show()
```

To be able to compare several countries, we may want to consider the country's population and compare the percentage of infected individuals in relation to the country's population. Let's load the countries dataset to get the population of each country:

 for our pandemic data. To do this, we will take a rolling window of 8 ninfected values, and apply the function to compute the ratio above:

```python
df['Rt'] = df['ninfected'].rolling(8).apply(lambda x: x[4:].sum()/x[:4].sum())
df['Rt'].plot()
plt.show()
```

There are some gaps in the graph, as you can see. These can be caused by either NaN or inf values in the dataset. Inf can be caused by division by zero, and NaN can be caused by missing data or not having enough data to compute the result (as in the beginning of our frame, where a rolling window of width 8 is not yet available). To improve the appearance of the graph, we must fill in the missing values with the replace and fillna functions.

Let's take a closer look at the start of the pandemic. In order to see better, we will also limit the y-axis values to only those below 6, and draw a horizontal line at 1.

```python
ax = df[df.index<"2020-05-01"]['Rt'].replace(np.inf,np.nan).fillna(method='pad').plot(figsize=(10,3))
ax.set_ylim([0,6])
ax.axhline(1,linestyle='--',color='black')
plt.show()
```
