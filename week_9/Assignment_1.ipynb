```python
from sklearn.preprocessing import LabelEncoder
new_columns = ['type', 'locale', 'locale_name', 'description', 'transferred']
events = event.drop([c for c in event.columns if c not in new_columns], axis=1)
events.dropna(inplace=True)
events = events.apply(LabelEncoder().fit_transform)
```

```python
events
```

```python
events['Time'] = np.arange(len(event.index))

events
```
