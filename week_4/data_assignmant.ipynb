```python
sns.relplot(data=pumpkins, x='High Price', y= 'Low Price')
plt.title('Relationship between High Price and Low Price')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.show()
```

```python


# Select the columns for the density plot
columns = ['Low Price', 'High Price']

# Create subplots for each column
fig, axes = plt.subplots(nrows=1, ncols=len(columns), figsize=(12, 4))

# Iterate over the columns and plot the density
for i, col in enumerate(columns):
    sns.kdeplot(data=pumpkins, x=col, ax=axes[i])
    axes[i].set_title(col)

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
